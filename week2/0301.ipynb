```python
import pandas as pd

df = pd.read_csv('class_1.csv')

df
```

```python
temp = df[['probe','pressure']]
temp
```

```python
Type = temp.groupby(['probe'])['pressure'].max().reset_index()
Type
```

```python
Type['class'] = 1
Type
```

```python
frames = [Type, Type2, Type3]
result = pd.concat(frames)
```
