```python
import openai
import pandas as pd
```

```python
keyfile = open("key.txt", "r")
key = keyfile.readline()
openai.api_key = key
```

```python
def chatgptfn(sub_list):
    result = ''
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an assistant"},
            {"role": "user", "content": f"{sub_list} :give me a summary"}
        ]
    )
    for choice in response.choices:
        result += choice.message.content
    return result
```

```python
Data = ['','','','','']
```

```python
for j in range(1,5):
    with open('0'+str(j)+'.txt', 'r', encoding="Latin-1") as fh:
        tmp = fh.read()
        itemlist = tmp.split(',')
    itemlist = str(itemlist)
    data = [itemlist[0:3200], itemlist[3201:6400], itemlist[6401:9600], itemlist[9601:12800], itemlist[12801:16000]]
    for i in range(0,5):
        data[i] = chatgptfn(data[i])
    Data[j] = ''
    for i in range(0,5):
        Data[j] = Data[j] + data[i]
```

```python
papers = [Data[1],Data[2],Data[3],Data[4]]
```

```python
import gensim
from gensim import corpora
from pprint import pprint
```

```python
# 創建詞袋
texts = [[word for word in document.lower().split()] for document in papers]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# 訓練 LDA 模型
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, 
                                            num_topics=3, random_state=100, update_every=1, 
                                            chunksize=100, passes=10, alpha='auto', per_word_topics=True)
```

```python
# 輸出主題模型分析結果
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
